```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# use local html as a practice
url = "https://mori-m-foundation.or.jp/wordpress/english/ius2"
html = requests.get(url)
soup = BeautifulSoup(html.content, "html.parser")

# blank list
research_titles = []

research = soup.find(id="research")
for title in research.find_all(class_="title"):
    research_titles.append(title.text)


# Make a dataframe
df = pd.DataFrame({"Research title": research_titles})

df.to_csv("ius_research.csv", encoding="utf_8_sig", index=False)

print("it's done!")


```
